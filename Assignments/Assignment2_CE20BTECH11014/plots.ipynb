```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
avl_file = "avl_output.txt"
bst_file = "bst_output.txt"

avl_data = {
    "Num_Comparisions":[],
    "Num_Operations":[],
    "Height":[]
}

bst_data = {
    "Num_Comparisions":[],
    "Height":[]
}
```

```python

with open(avl_file) as file:
    for lines in file:
        data = lines.split(" ")
        avl_data["Num_Comparisions"].append(int(data[0]))
        avl_data["Num_Operations"].append(int(data[1]))
        avl_data["Height"].append(int(data[2]))
    avl_data = pd.DataFrame(avl_data)

with open(bst_file) as file:
    for lines in file:
        data = lines.split(" ")
        bst_data["Num_Comparisions"].append(int(data[0]))
        bst_data["Height"].append(int(data[1]))
    bst_data = pd.DataFrame(bst_data)
```

```python
y = bst_data["Height"].to_list()
average_height_bst = 0
for i in y:
    average_height_bst+=i/100
normal = {}
for heights in set(y):
    normal[heights] = y.count(heights)
normal_key = list(normal.keys())
normal_key.sort()
normal_sorted_key = {i:normal[i] for i in normal_key}

plt.bar(normal_sorted_key.keys(), normal_sorted_key.values())
plt.title("Normal Distribution of height of trees")
plt.ylabel("Number of Trees")
plt.xlabel("Height")
plt.show()
patches,text = plt.pie(normal_sorted_key.values())
plt.legend(patches,normal_key,loc="upper right")
plt.axis("equal")
plt.title("Distribution of height of trees")
plt.show()
```

```python
y = bst_data["Num_Comparisions"].to_list()
average_num_comparisions_bst = 0
for i in y:
    average_num_comparisions_bst+=int(i)/100
normal = {}
for comp in set(y):
    normal[comp] = y.count(comp)
normal_key = list(normal.keys())
normal_key.sort()
normal_sorted_key = {i:normal[i] for i in normal_key}
print(average_num_comparisions_bst)
print(average_height_bst)
```

```python
y = avl_data["Height"].to_list()
average_height_avl = 0
for i in y:
    average_height_avl+=i/100
normal = {}
for heights in set(y):
    normal[heights] = y.count(heights)
normal_key = list(normal.keys())
normal_key.sort()
normal_sorted_key = {i:normal[i] for i in normal_key}

patches, text = plt.pie(normal_sorted_key.values())
plt.legend(patches,normal_key,loc="upper right")
plt.title("Distribution of height of trees")
plt.show()
```

```python
y = avl_data["Num_Comparisions"].to_list()

average_num_comparisions_avl = 0
average_num_operations_avl = 0
for i in y:
    average_num_comparisions_avl+=int(i)/100
y = avl_data["Num_Operations"]
for i in y:
    average_num_operations_avl+=int(i)/100
total_average_operations_avl = average_num_comparisions_avl+average_num_operations_avl
print(average_num_comparisions_avl)
print(average_num_operations_avl)
print(total_average_operations_avl)
print(average_height_avl)
```

```python

result = {
    "BST" : [average_num_comparisions_bst,average_num_comparisions_bst],
    "AVL" : [average_num_comparisions_avl,total_average_operations_avl]
}
heights_result = {
    "BST" : average_height_bst,
    "AVL" : average_height_avl
}

df = pd.DataFrame(result,columns=["BST","AVL"], index=["average_comparisions","total_average_operations"])
heights_df = pd.DataFrame(heights_result,columns=["BST","AVL"], index=["average_height"])

df.plot.barh()
plt.show()
heights_df.plot.barh()
plt.show()
```
